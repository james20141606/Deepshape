```python
import pandas as pd
```

```python
rbp_names = pd.read_table('metadata/ionmf.rbp.txt', header=None, names=['index', 'name'])
rbp_names.index = rbp_names['index']
rbp_names = rbp_names['name']
```

```python
summary = pd.read_table('reports/ionmf_motif/summarize_metrics.txt')
summary['dataset'] = summary['dataset'].apply(lambda x: x.split('_')[0]).astype('int')
summary['dataset'] = rbp_names.values[summary['dataset'].values]
summary['method'] = summary['icshape_dataset'] + '.' + summary['model_name'] + '.' + summary['featureset']
summary.head()
```

```python
roc_auc_other_methods = pd.read_table('reports/ionmf_motif/roc_auc_table.other_methods.txt')
roc_auc_other_methods.index = roc_auc_other_methods['rbp_name']
roc_auc_other_methods.head()
```

```python
roc_auc_table = summary.pivot_table(['roc_auc'], index=['dataset'], columns=['method'])
roc_auc_table.columns = roc_auc_table.columns.droplevel(0)
roc_auc_table_total = pd.merge(roc_auc_table, roc_auc_other_methods, left_index=True, right_index=True)
roc_auc_table_total.drop(['rbp_name', 'index'], axis=1, inplace=True)
roc_auc_table_total.index.name = 'rbp_name'
roc_auc_table_total.to_csv('reports/ionmf_motif/roc_auc_table.txt', sep='\t')
roc_auc_table_total
```
